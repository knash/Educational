```python
#Test
from __future__ import print_function
import argparse
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import sys
from torchvision import datasets, transforms
from torch.optim.lr_scheduler import StepLR
from d2l import torch as d2l
import torch.nn.functional as F
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
#inputs X, Weights W, bias b

batch_size=128
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
n_ep=30

def relu(X):
    return(torch.max(X,torch.zeros_like(X)))
class MLP(nn.Module):
    def __init__(self,num_inputs, num_outputs, num_hiddens):
        super().__init__()
        self.W1=nn.Parameter(torch.randn(num_inputs,num_hiddens),requires_grad=True)
        self.W2=nn.Parameter(torch.randn(num_hiddens,num_outputs),requires_grad=True)
        self.b1=nn.Parameter(torch.randn(num_hiddens),requires_grad=True)
        self.b2=nn.Parameter(torch.randn(num_outputs),requires_grad=True)
    def forward(self,X):
        X=X.reshape(-1,num_inputs)
        H1=relu(X@self.W1+self.b1)
        return H1@self.W2+self.b2
lossfunc=nn.CrossEntropyLoss()
losses=[]
lossestest=[]
num_inputs, num_outputs, num_hiddens = 784, 10, 16
MLPmod=MLP(num_inputs, num_outputs, num_hiddens)

Aopt=torch.optim.Adam(MLPmod.parameters(),lr=0.01)

for iep in range(n_ep):
    lossep=0.
    nep=0
    for x,y in train_iter:
        Aopt.zero_grad()
        loss=lossfunc(MLPmod(x),y).mean()
        print (loss)
        loss.backward()
        Aopt.step()
        lossep+=loss
        nep+=1
    losses.append((lossep/nep).detach().numpy())
    losseptest=0.
    nep=0
    for x,y in test_iter:
        losseptest+=lossfunc(MLPmod(x),y).mean()
        nep+=1
    lossestest.append((losseptest/nep).detach().numpy())

    clear_output(wait=True)

    plt.plot(losses)
    plt.plot(lossestest)
    plt.yscale("log")

    plt.pause(0.0001)
```
