```python
#HDF5 is for storing large scientific arrays
#hierarchichal data format. supports various compression algorithms 
#hdfstore creates a file then accesses with dict-like syntax
#hdfstore is good for write once read many applications

store=pd.HDFStore("hdftype.hdf")
store["df2"]=df2
print(store["df2"])

#put method does this above and allows options
#this part seems stupid though
store.put("df2",df2,format="table")
print(store.select("df2",where =["index>15"]))

#you can also just use the usual
df2.to_hdf("df2.h5","df2",format="fixed")
```
